```python
import random
import pandas as pd
import matplotlib.pyplot as plt

computer_wins = 0
user_wins = 0
rounds = int(input("Enter how many times you want to play this game: "))
results = []

for i in range(0, rounds):
    computer = random.randint(1, 9)
    userinput = int(input("Enter the number between 1 and 9: "))
    print("Computer:", computer)
    print("User input:", userinput)

    if computer > userinput:
        print("Computer won")
        computer_wins += 1
        results.append("Computer")
    elif computer == userinput:
        print("Tie")
        results.append("Tie")
    else:
        print("You won")
        user_wins += 1
        results.append("User")

print("Score of computer:", computer_wins)
print("Your score:", user_wins)

# Save results to a CSV file
data = {'Round': range(1, rounds+1), 'Result': results}
df = pd.DataFrame(data)
df.to_csv('lottery_results.csv', index=False)

# Plotting the success rate
total_games = rounds
user_success_rate = user_wins / total_games * 100
computer_success_rate = computer_wins / total_games * 100

plt.bar(['User', 'Computer'], [user_success_rate, computer_success_rate], color=['blue', 'orange'])
plt.xlabel('Player')
plt.ylabel('Success Rate (%)')
plt.title('Success Rate of User vs. Computer')
plt.ylim(0, 100)
plt.show()
```
